# Level Data

```python
Z             = []
Ion           = []
E             = []
Level         = []
N             = []
L             = []
S             = []
J             = []
Energy        = []
Potential     = []
Configuration = []

for n in range(390627):
    if Pointer[10*n+1]==6:
        Z             += [Integer[Pointer[10*n+8]+2]]
        Ion           += [Integer[Pointer[10*n+8]+4]]
        E             += [Electron[Integer[Pointer[10*n+8]+4]]]
        Level         += [Integer[Pointer[10*n+8]+3]]
        N             += [Integer[Pointer[10*n+8]-1]]
        L             += [Integer[Pointer[10*n+8]+1]]
        S             += [Integer[Pointer[10*n+8]+0]]
        J             += [int(Real[Pointer[10*n+7]+0])]
        Energy        += ["{0:.6e}".format(Real[Pointer[10*n+7]-1])]
        Potential     += ["{0:.6e}".format(Real[Pointer[10*n+7]+2])]
        Configuration += ["".join(list(map(chr,String[Pointer[10*n+9]-1:Pointer[10*n+19]-1])))]

df01 = DataFrame(data={"Z":Z, "Ion":Ion, "Electron":E, "Level":Level, "N":N, "L":L, "2S+1":S, "2J+1":J, "Energy":Energy, "Potential":Potential, "Configuration":Configuration})
df01
```

```python
for n in range(92,106):
    df01.query("Ion=={0:}".format(n)).to_csv("./database/xstar/Si/Si{0:02d}_Level.tsv".format(106-n),sep="\t")
for n in range(326,352):
    df01.query("Ion=={0:}".format(n)).to_csv("./database/xstar/Fe/Fe{0:02d}_Level.tsv".format(352-n),sep="\t")
```

# Bound-bound Radiative

```python
Z             = []
Ion           = []
Initial_Level = []
Final_Level   = []
Energy        = []
Strength      = []
A_coefficient = []

for n in range(390628):
    if Pointer[10*n+1]==50:
        if Real[Pointer[10*n+7]-1]!=0:
            Z             += [Integer[Pointer[10*n+8]+1]]
            Ion           += [Integer[Pointer[10*n+8]+2]]
            Initial_Level += [Integer[Pointer[10*n+8]+0]]
            Final_Level   += [Integer[Pointer[10*n+8]-1]]
            Energy        += ["{0:.6e}".format(12398/Real[Pointer[10*n+7]-1])]
            Strength      += ["{0:.6e}".format(      Real[Pointer[10*n+7]+0])]
            A_coefficient += ["{0:.6e}".format(      Real[Pointer[10*n+7]+1])]

df03 = DataFrame(data={"Z":Z, "Ion":Ion, "Initial Level":Initial_Level, "Final Level":Final_Level, "Energy":Energy, "Strength":Strength, "Rate":A_coefficient})
df03 = df03.sort_values(["Ion","Initial Level","Final Level"]).reset_index().drop(columns="index")
df03
```

```python
for n in range(92,106):
    df03.query("Ion=={0:}".format(n)).to_csv("./database/xstar/Si/Si{0:02d}_Transition.tsv".format(106-n),sep="\t")
```

# Superlevel Photoionization

```python
Z             = []
Initial_Ion   = []
Initial_Level = []
Final_Ion     = []
Final_Level   = []
N             = []
L             = []
S             = []
Energy        = []
Cross         = []

for i in range(390628):
    if Pointer[10*i+1]==70 or Pointer[10*i+1]==99:
        nd = Integer[Pointer[10*i+8]-1]
        nt = Integer[Pointer[10*i+8]+0]
        nx = Integer[Pointer[10*i+8]+1]

        for j in range(nx):
            Z             += [Integer[Pointer[10*i+8]+5]]
            Initial_Ion   += [Integer[Pointer[10*i+8]+9]]
            Initial_Level += [Integer[Pointer[10*i+8]+8]]
            Final_Ion     += [Integer[Pointer[10*i+8]+7]]
            Final_Level   += [Integer[Pointer[10*i+8]+6]]
            N             += [Integer[Pointer[10*i+8]+2]]
            L             += [Integer[Pointer[10*i+8]+3]]
            S             += [Integer[Pointer[10*i+8]+4]]
            Energy        += ["{0:.6e}".format(Real[Pointer[10*i+7]+nd+nt+nd*nt+2*j-1])]
            Cross         += ["{0:.6e}".format(Real[Pointer[10*i+7]+nd+nt+nd*nt+2*j+0])]

df04 = DataFrame(data={"Z":Z, "Initial_Ion":Initial_Ion, "Initial Level":Initial_Level, "Final_Ion":Final_Ion, "Final Level":Final_Level, "N":N, "L":L, "2S+1":S, "Energy (Ry)":Energy, "Cross Section (Mb)":Cross})
df04 = df04.sort_values(["Initial_Ion", "Initial Level", "Final Level"]).reset_index().drop(columns="index")
df04
```

```python
for n in range(92,106):
    df04.query("Final_Ion=={0:}".format(n)).to_csv("./database/xstar/Si/Si{0:02d}_Photoionization.tsv".format(106-n),sep="\t")
```

# Superlevel Recombination

```python
Z             = []
Initial_Ion   = []
Initial_Level = []
Final_Ion     = []
Final_Level   = []
N             = []
L             = []
S             = []
Temperature   = []
Rate          = []

for i in range(390628):
    if Pointer[10*i+1]==70:
        nd = Integer[Pointer[10*i+8]-1]
        nt = Integer[Pointer[10*i+8]+0]
        nx = Integer[Pointer[10*i+8]+1]
        
        for j in range(nt):
            Z             += [Integer[Pointer[10*i+8]+5]]
            Initial_Ion   += [Integer[Pointer[10*i+8]+7]]
            Initial_Level += [Integer[Pointer[10*i+8]+6]]
            Final_Ion     += [Integer[Pointer[10*i+8]+9]]
            Final_Level   += [Integer[Pointer[10*i+8]+8]]
            N             += [Integer[Pointer[10*i+8]+2]]
            L             += [Integer[Pointer[10*i+8]+3]]
            S             += [Integer[Pointer[10*i+8]+4]]
            Temperature   += ["{0:.6e}".format(10**Real[Pointer[10*i+7]+nd+j-1])]
            Rate          += ["{0:.6e}".format(10**Real[Pointer[10*i+7]+nd+nt+j-1])]

df05 = DataFrame(data={"Z":Z, "Initial_Ion":Initial_Ion, "Initial Level":Initial_Level, "Final_Ion":Final_Ion, "Final Level":Final_Level, "N":N, "L":L, "2S+1":S, "Temperature":Temperature, "Rate":Rate})
df05 = df05.sort_values(["Z", "Initial_Ion", "Initial Level", "Final_Ion"]).reset_index().drop(columns="index")
df05
```
